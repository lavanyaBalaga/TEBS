# Structuring the data according to the needs

```python
import pandas as pd
df=pd.read_csv("/content/non_purchased_users.csv")
```

```python
df
```

```python
import ast
df['unifiedScreenName'] = df['unifiedScreenName'].apply(ast.literal_eval)
df['userEngagementDuration'] = df['userEngagementDuration'].apply(ast.literal_eval)

# Expand the lists into separate rows
df_expanded = df.explode(['unifiedScreenName', 'userEngagementDuration'])

# Convert the date column to datetime format
df_expanded['date'] = pd.to_datetime(df_expanded['date'], format='%d-%m-%Y')

# Pivot the table
pivot_table = df_expanded.pivot_table(index=['customUser:user_Id', 'date'],
                                      columns='unifiedScreenName',
                                      values='userEngagementDuration',
                                      aggfunc='first').fillna(0)

# Reset index to flatten the DataFrame
pivot_table = pivot_table.reset_index()

# Save to CSV (if needed)
pivot_table.to_csv('pivot_table.csv', index=False)

print(pivot_table)
```
