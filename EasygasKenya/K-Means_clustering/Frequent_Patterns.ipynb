```python
def fetch_ga4_data():
    client = BetaAnalyticsDataClient()
    request = RunReportRequest(
        property="286596396",
        dimensions=[Dimension(name="userType")],
        metrics=[Metric(name="Users")],
        date_ranges=[DateRange(start_date="2024-01-01", end_date="2024-05-15")]
    )
    response = client.run_report(request)

    data = []
    for row in response.rows:
        record = {}
        for dimension, value in zip(response.dimension_headers, row.dimension_values):
            record[dimension.name] = value.value
        for metric, value in zip(response.metric_headers, row.metric_values):
            record[metric.name] = value.value
        data.append(record)

    df = pd.DataFrame(data)
    return df

# Function to classify users
def classify_users(df):
    df['user_status'] = df['activeUsers'].apply(lambda x: 'active' if x > 10 else 'non-active')
    # Add more classifications as needed
    return df

# Fetch and classify data
df = fetch_ga4_data()
df = classify_users(df)

# Save the classified data to an Excel file
df.to_excel("classified_users.xlsx", index=False)

# Display the DataFrame
df.head()
```
