```python
df=pd.read_csv("/content/Recent_with_screens.csv")
df.head(10)
```

```python
# Filter the dataframe to get pur_users who visited '/ChoosePayment'
pur_users = df[df['unifiedScreenName'] == '/ChoosePayment']

# Get the count of pur_users
pur_users_count = pur_users['customUser:user_Id'].nunique()
pur_user_ids = pur_users['customUser:user_Id'].unique()
print(pur_users_count)
# Filter the dataframe to include only the pur_users and their visited pages along with engagement duration
pur_users_data = df[df['customUser:user_Id'].isin(pur_user_ids)][['customUser:user_Id', 'unifiedScreenName', 'userEngagementDuration']]

```

```python
# Group the data by user ID and list the unified screen names along with their engagement time
pur_users_grouped = pur_users_data.groupby('customUser:user_Id').agg({
    'unifiedScreenName': lambda x: list(x),
    'userEngagementDuration': lambda x: list(x)
}).reset_index()

pur_users_grouped.head(20)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Prepare data for clustering
# Create a pivot table with user_ids as rows, screen names as columns, and engagement durations as values
pivot_table = pur_users_data.pivot_table(index='customUser:user_Id', columns='unifiedScreenName', values='userEngagementDuration', aggfunc=np.sum).fillna(0)

# Standardize the data
scaler = StandardScaler()
pivot_table_scaled = scaler.fit_transform(pivot_table)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Assuming we are looking for 3 clusters
clusters = kmeans.fit_predict(pivot_table_scaled)

# Add the cluster labels to the original data
pur_users_grouped['cluster'] = clusters

# Display the clustered data
pur_users_grouped_sorted = pur_users_grouped.sort_values(by='cluster')
pur_users_grouped_sorted
```

```python
# Filter the previously identified pur_users data for the selected screens
selected_screens = ['/ProductListing', '/CartScreen', '/ChoosePayment', '/LoginScreen', '/OtpScreen', '/ProductDetails', '/DashboardScreen']
pur_users_filtered = pur_users[pur_users['unifiedScreenName'].isin(selected_screens)]

# Aggregate the data for each pur_user
pur_users_grouped_filtered = pur_users_filtered.groupby('customUser:user_Id').agg({'unifiedScreenName': lambda x: list(x),'userEngagementDuration': lambda x: list(x)}).reset_index()

pur_users_grouped_filtered
```

```python
import pandas as pd
from collections import defaultdict
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load data
file_path = '/content/pur_users_grouped_extended_filtered.csv'
df = pd.read_csv(file_path)

# Group data by customer_id and aggregate screen_name and duration
df['userEngagementDuration'] = df.apply(lambda x: (x['unifiedScreenName'], x['userEngagementDuration']), axis=1)
grouped = df.groupby('customUser:user_Id')['userEngagementDuration'].apply(list).reset_index()

# Convert sequences to a format suitable for sequence mining
sequences = grouped['userEngagementDuration'].tolist()

# Extract just the screen names for frequent itemset mining
screen_sequences = [[screen for screen, duration in sequence] for sequence in sequences]

# Use TransactionEncoder to transform the data
te = TransactionEncoder()
te_ary = te.fit(screen_sequences).transform(screen_sequences)
df_transformed = pd.DataFrame(te_ary, columns=te.columns_)

# Apply the Apriori algorithm
frequent_itemsets = apriori(df_transformed, min_support=0.1, use_colnames=True)

# Find the association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

# Print the frequent itemsets and rules
print(frequent_itemsets)
#print(rules)

# Prepare a user-pattern dictionary
user_patterns = defaultdict(list)

# Identify users associated with each pattern
for index, row in grouped.iterrows():
    customer_id = row['customUser:user_Id']
    sequence = row['userEngagementDuration']
    screen_names = [screen for screen, duration in sequence]

    for itemset in frequent_itemsets['itemsets']:
        if itemset.issubset(screen_names):
            user_patterns[customer_id].append(list(itemset))

# Print user patterns
'''for user, patterns in user_patterns.items():
    print(f"Customer ID: {user}, Patterns: {patterns}")'''
```
