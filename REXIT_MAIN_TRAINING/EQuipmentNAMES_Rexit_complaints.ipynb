```python
import nltk
from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import pandas as pd

# Sample list of sentences
df=pd.read_csv("/content/test_rexit.csv")

# Tokenize the sentences
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()

equipment_names_list = []

# Get the NLTK English stop words list
stop_words = set(stopwords.words('english'))

for sentence in df['sentences']:
    tokens = tokenizer.tokenize(sentence.lower())  # Convert to lowercase for case-insensitive matching
    pos_tags = pos_tag(tokens)

    equipment_names = set()

    for i in range(len(pos_tags)):
        if pos_tags[i][1] == 'NN' or pos_tags[i][1] == 'NNS':  # Check if the word is a noun
            if i > 0 and pos_tags[i-1][1] == 'JJ':  # Check if the previous word is an adjective
                equipment_names.add(pos_tags[i-1][0] + ' ' + pos_tags[i][0])
            else:
                # Lemmatize the word to its base form
                base_word = lemmatizer.lemmatize(pos_tags[i][0])
                if base_word not in stop_words:  # Check if the word is not a stop word
                    equipment_names.add(base_word)

    # Append equipment names for this sentence to the list
    equipment_names_list.append(equipment_names)

# Add a new column to the DataFrame containing the extracted equipment names
df['equipment_names'] = equipment_names_list

df
```

```python
df=pd.read_csv("/content/generated.csv")
lemmatizer = WordNetLemmatizer()

equipment_names_list = []

# Define a function to convert the Penn Treebank POS tags to WordNet POS tags
def penn_to_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

for sentence in df['Description']:
    # Tokenize the sentence
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(sentence.lower())  # Convert to lowercase for case-insensitive matching

    # Perform part-of-speech tagging using NLTK's POS tagger
    pos_tags = pos_tag(tokens)

    equipment_names = set()

    for word, pos in pos_tags:
        # Convert the Penn Treebank POS tag to WordNet POS tag
        wn_pos = penn_to_wordnet_pos(pos)

        if wn_pos == wordnet.NOUN or wn_pos == wordnet.NOUN:  # Check if the word is a noun or adjective
            # Lemmatize the word to its base form using WordNet Lemmatizer
            base_word = lemmatizer.lemmatize(word, pos=wn_pos)
            equipment_names.add(base_word)

    # Append equipment names for this sentence to the list
    equipment_names_list.append(equipment_names)

# Add a new column to the DataFrame containing the extracted equipment names
df['equipment_names'] = equipment_names_list
df
```

```python
import nltk

from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
from nltk.stem import WordNetLemmatizer
import pandas as pd

df=pd.read_csv("/content/REXITDATATILLAPRIL1.csv")

# Tokenize the sentences
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
equipment_names_list = []
for sentence in df['Description']:
    if pd.notna(sentence):  # Check if the value is not NaN
        tokens = tokenizer.tokenize(sentence.lower())  # Convert to lowercase for case-insensitive matching
        pos_tags = pos_tag(tokens)

        # Perform Named Entity Recognition (NER)
        named_entities = ne_chunk(pos_tags)
        equipment_names = set()
        for entity in named_entities:
            if isinstance(entity, tuple):  # If it's a single word
                word, pos = entity
                if pos == 'NN' or pos == 'NNS':  # Check if the word is a noun
                    base_word = lemmatizer.lemmatize(word)
                    equipment_names.add(base_word)
            elif hasattr(entity, 'label') and entity.label() != 'PERSON' and entity.label()!='NAME':  # If it's a named entity other than a person
                entity_words = [word for word, pos in entity.leaves() if pos == 'NN' or pos == 'NNS']  # Extract nouns from the entity
                entity_words = [lemmatizer.lemmatize(word) for word in entity_words]  # Lemmatize the words
                equipment_names.update(entity_words)
        equipment_names_list.append(equipment_names)
    else:
        equipment_names_list.append(set())  # Append an empty set for NaN values
# Add a new column to the DataFrame containing the extracted equipment names
df['Equipment Names'] = equipment_names_list

df1=df[['NMD','Description','Equipment Names']]
df1.tail(25)
```
