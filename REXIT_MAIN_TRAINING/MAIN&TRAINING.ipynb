# Supervised-Learning methods for maintenance and training classification

For rexit analytics,after getting the equipment names, next thing is to get the classification where a complaint is "maintenance-related" or "training-related".

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
import pandas as pd
df=pd.read_csv("/content/test2.csv")
# Split data into features (X) and labels (y)
X = df['Description']
y = df['Area']

# Build a pipeline
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(X, y)

# Classify a new sentence
new_sentence = "Traffic warden found guiding the 3P vehicles without flags and 3P vehicle driver was not wearing the seat belt. Stop card shown to traffic warden and 3P driver immediately."
predicted_category = model.predict([new_sentence])[0]
print(f"The predicted category for '{new_sentence}' is: {predicted_category}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Example complaint data and their corresponding labels (1 for maintenance, 0 for training)
complaints = [
    "The circuit breaker tripped again.",  # Maintenance
    "We need to schedule a training session for new employees.",  # Training
    "Dispose of the waste properly.",  # Maintenance
    "The equipment needs repair.",  # Maintenance
    "Let's discuss the latest safety training procedures.",  # Training
    "There are cracks in the wall.",  # Maintenance
    "Our team is undergoing maintenance training next week."  # Training
]

labels = [1, 0, 1, 1, 0, 1, 0]

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Vectorize complaints
X = vectorizer.fit_transform(complaints)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Train logistic regression classifier
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict labels for test data
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
unlabelled_texts = df_unlabelled['Description'].tolist()

# Tokenize the unlabelled data
unlabelled_encodings = tokenizer(unlabelled_texts, truncation=True, padding=True, return_tensors='pt')

# Create a dataset for the unlabelled data
class UnlabelledDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        item = {key: val[idx].clone().detach() for key, val in self.encodings.items()}
        return item

    def __len__(self):
        return len(self.encodings.input_ids)

unlabelled_dataset = UnlabelledDataset(unlabelled_encodings)

# Use DataLoader for batching
unlabelled_dataloader = torch.utils.data.DataLoader(unlabelled_dataset, batch_size=8)

# Evaluate and make predictions
predictions = []
model.eval()
for batch in unlabelled_dataloader:
    with torch.no_grad():
        outputs = model(**batch)
    predictions.extend(torch.argmax(outputs.logits, dim=1).tolist())

# Add predictions to DataFrame
df_unlabelled['data_label'] = predictions

# Map predictions to corresponding labels
label_map = {0: 'maintenance', 1: 'training'}
df_unlabelled['data_label'] = df_unlabelled['data_label'].map(label_map)

# Display the DataFrame with the new column
print(df_unlabelled)
```
